### Code

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
nik = ["0882","0853","0891"]
jeneng = ["Anes","Prbawo","Gahar"]
hobi = ["Makan","Bernafas","Melihat"]
for i,data_jeneng in enumerate(jeneng):
  print(f"{nik[i]} ... {data_jeneng}")

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
for data_nik, data_jeneng, data_hobi in zip(nik,jeneng,hobi):
  print(f"{data_nik} -- {data_jeneng} -- {data_hobi}")
```

### Code

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
def dilipat(x):
  return x * 2

input = [10,20,30]
output = []
for item in input:
  output.append(dilipat(item))
print(f"Input {input}")
print(f"Output {output}")

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
output = list(map(dilipat, input))
print(f"Output {output}")
```

```python
# Cara Tidak Pythonic
print("Cara Tidak Pythonic")
input = [30,55,60,90,50]
output = []
for i in input:
  if i > 50:
    output.append(i)
print(output)

# Cara Lebih Pythonic
print("\nCara Lebih Pythonic")
output = [i for i in input if i > 50]
print(output)
```
